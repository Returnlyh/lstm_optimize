```python
def CreateDictFromFlatbuffer(buffer_data):
    model_obj = schema_fb.Model.GetRootAsModel(buffer_data, 0)
    model = schema_fb.ModelT.InitFromObj(model_obj)
    return FlatbufferToDict(model, preserve_as_numpy=False)
```

```python
def FlatbufferToDict(fb, preserve_as_numpy, print_ = ">>>"):
    if isinstance(fb, int) or isinstance(fb, float) or isinstance(fb, str):
        return fb
    elif hasattr(fb, "__dict__"):
        result = {}
        for attribute_name in dir(fb):
            attribute = fb.__getattribute__(attribute_name)
            if not callable(attribute) and attribute_name[0] != "_":
                snake_name = CamelCaseToSnakeCase(attribute_name)
                preserve = True if attribute_name == "buffers" else preserve_as_numpy
                print(f"{print_}:{snake_name}")
                result[snake_name] = FlatbufferToDict(attribute, preserve, print_=f"{print_}>>>")
        return result
    elif isinstance(fb, np.ndarray):
        return fb if preserve_as_numpy else fb.tolist()
    elif hasattr(fb, "__len__"):
        return [FlatbufferToDict(entry, preserve_as_numpy) for entry in fb]
    else:
        return fb
```

```python
def CamelCaseToSnakeCase(camel_case_input):
    """Converts an identifier in CamelCase to snake_case."""
    s1 = re.sub("(.)([A-Z][a-z]+)", r"\1_\2", camel_case_input)
    return re.sub("([a-z0-9])([A-Z])", r"\1_\2", s1).lower()
```

```python
import tensorflow as tf
import numpy as np

from tensorflow.keras import Sequential
from tensorflow.keras.layers import LSTM
```

```python
units = 4
model = Sequential()
model.add(LSTM(units, input_shape=(2, 8), activation='tanh',recurrent_activation='sigmoid', return_sequences=True))
```

```python
print(model.get_weights()[0].shape)
```

```python
x = np.random.randn(10, 2, 8)
y = model.predict(x)
print(y.shape, y)
```

```python
w_x = model.get_weights()[0]
w_h = model.get_weights()[1]
bias = model.get_weights()[2]
```

```python
# 4.1获取输⼊⻔参数
w_ix = w_x[:,:units] #输⼊⻔控中，输⼊层和循环层的权重
w_ih = w_h[:,:units] # 输⼊⻔控中，上⼀时刻循环层和当前的循环层之间的权重
b_i = bias[:units] #输⼊⻔控中，当前时刻循环层的偏置
print(f"{w_ix.shape}, {w_ih.shape}, {b_i.shape}")
```

```python
# 4.2获取遗忘⻔参数
w_fx = w_x[:,units:units*2] #遗忘⻔控中，输⼊层和循环层的权重
w_fh = w_h[:,units:units*2] # 遗忘⻔控中，上⼀时刻循环层和当前的循环层之间的权重
b_f = bias[units:units*2] #遗忘⻔控中，当前时刻循环层的偏置
print(f"{w_fx.shape}, {w_fh.shape}, {b_f.shape}")
```

```python
# 4.3获取当前输⼊状态参数
w_cx = w_x[:,2*units:units*3] #当前输⼊状态，输⼊层和循环层的权重
w_ch = w_h[:,2*units:units*3] # 当前输⼊状态，上⼀时刻循环层和当前的循环层之间的权重
b_c = bias[2*units:units*3] #当前输⼊状态，当前时刻循环层的偏置
print(f"{w_cx.shape}, {w_ch.shape}, {b_c.shape}")
```

```python
# 4.4获取输出⻔参数
w_ox = w_x[:,3*units:units*4] #输出⻔控中，输⼊层和循环层的权重
w_oh = w_h[:,3*units:units*4] # 输出⻔控中，上⼀时刻循环层和当前的循环层之间的权重
b_o = bias[3*units:units*4] #输出⻔控中，当前时刻循环层的偏置
print(f"{w_ox.shape}, {w_oh.shape}, {b_o.shape}")
```

```python
# 5.激活函数sigmoid实现
def activation_sigmoid(x):
    for i in range(x.shape[0]):
        for j in range(x.shape[1]):
            x[i,j] = 1 / (1 + np.exp(-x[i,j]))
    return x

def activation_tanh(x):
    for i in range(x.shape[0]):
        for j in range(x.shape[1]):
            x[i,j] = (np.exp(x[i,j]) - np.exp(-x[i,j])) / (np.exp(x[i,j]) + np.exp(-x[i,j]))
    return x
```

```python
# 6.⼿动实现的前向推理，因为只有⼀个时间步，所以没有h_{t-1}， h_t为lstm的输出
x_ = x[:,0,:]
f_t = activation_sigmoid(np.dot(x_,w_fx)+b_f)
i_t = activation_sigmoid(np.dot(x_,w_ix)+b_i)
c_hat_t = activation_tanh(np.dot(x_,w_cx)+b_c)
o_t = activation_sigmoid(np.dot(x_,w_ox)+b_o)
c_t = np.multiply(i_t,c_hat_t)
h_t = np.multiply(o_t,activation_tanh(c_t))
print(f_t.shape, i_t.shape, c_hat_t.shape, o_t.shape, c_t.shape, h_t.shape)
print(f"[INFO]>>> ⼿动前向计算输出:{h_t}")
```

```python
# 7.⽐较⼿动运算的输出h_t和模型预测的输出model_predict
def compare_value(a,b):
    if len(a.shape) != len(b.shape):
        print("两者维度不同，⽆法⽐较")
    else:
        a_shape = np.array(a.shape)
        b_shape = np.array(b.shape)
        result = a_shape-b_shape
        if sum(result) != 0:
            print("维度不同，⽆法⽐较")
        else:
            #为防⽌保留位数的差异，统⼀保留⼩数点后6位
            a = np.round(a,6)
            b = np.round(b,6)
            c = a - b
            #由于电脑计算位数等等⼀些⽅式，可能保留的⼩数位不同，会在⼩数点后6,7位有点点误差
            #这个不是运算⽅式，⽽是电脑保留位数导致的
            if c.sum() < 1e-5:
                print("两者相同")
            else:
                print("两者数据不同")
compare_value(h_t,y)
```
